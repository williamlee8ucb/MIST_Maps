```python
c = ComboTile('79 N / 91 W', 'Arctic_Data', center = [-90.74778, 79.41861])
```

```python
max_elev, max_r, max_z, max_coords = c.max_per_bin_fast(bin_width = 0.1)
```

```python
c.plot_max_per_bin(bin_width = 0.1)
```

```python
bin_width = 0.1
nbins = len(max_elev)
azimuths = np.arange(nbins) * bin_width
lons = max_coords[:, 0]
lats = max_coords[:, 1]

df = pd.DataFrame({"azimuth_angle_bins":azimuths,
                  "max_elevation_angles (deg)":max_elev,
                  "radial_distances (m)":max_r,
                  "heights (m)":max_z,
                  "longitude":lons,
                  "latitude":lats})
```
